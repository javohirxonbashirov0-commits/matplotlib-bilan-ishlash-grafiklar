```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
import zipfile
with zipfile.ZipFile("/content/archive (5).zip", "r") as z:
    with z.open("Expanded_data_with_more_features.csv") as f:
        df = pd.read_csv(f)
```

```python
df.head()
```

```python
df.columns
```

## Gender boyicha nisbat (pie chart)

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=140)
plt.title("Talabalarning jins bo'yicha nisbati")
plt.show()
```

# Pie chart yordamida ota - onalarining oilaviy holati bo'yicha

```python
marital_counts = df['ParentMaritalStatus'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(marital_counts, labels=marital_counts.index, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff','#99ff99','#ffcc99'])
plt.title("Ota-onalarning oilaviy holati bo'yicha taqsimoti")
plt.show()
```

#Barchart yordamida talablar soni bo'yicha  matematika ballari boyicha taqsimot

```python
plt.figure(figsize=(10, 6))
plt.hist(df['MathScore'], bins=20, color='red', edgecolor='black')
plt.title('Matematika ballari taqsimoti')
plt.xlabel('Ballar')
plt.ylabel('Talabalar soni')
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.show()
```

```python
parent_edu_scores = df.groupby('ParentEduc')['MathScore'].mean().sort_values()

plt.figure(figsize=(12, 6))
parent_edu_scores.plot(kind='barh', color='blue')
plt.title("Ota-onaning ma'lumoti bo'yicha o'rtacha matematika ballari")

plt.xlim(0, 100)
plt.xlabel('O‘rtacha ball')
plt.ylabel('Ota-ona ma’lumoti')
plt.show()
```

#scaterplot

```python
plt.figure(figsize=(8, 8))
plt.scatter(df['ReadingScore'], df['WritingScore'], alpha=0.5, color='green')
plt.title('O‘qish va Yozish ballari o‘rtasidagi bog‘liqlik')
plt.xlabel('O‘qish ballari')
plt.ylabel('Yozish ballari')
plt.plot([0, 100], [0, 100], color='red', linestyle='--')
plt.show()
```

#boxplot

```python
plt.figure(figsize=(8, 6))
df.boxplot(column='MathScore', by='Gender', patch_artist=True)
plt.title('Jinslar bo‘yicha  ballari')
plt.suptitle('')

plt.ylabel('Ballar')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.boxplot([df['MathScore'], df['ReadingScore'], df['WritingScore']],
            labels=['Matematika', 'O‘qish', 'Yozish',],
            patch_artist=True,
            boxprops=dict(facecolor="lightpink"))

plt.title("Ballar taqsimoti va Outlierlar")
plt.ylabel("Ballar")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.show()
```

Ota-onalar ma'lumoti va oilaviy holati bo'yicha taqsimot


```python
pivot_df = df.groupby(['ParentEduc', 'ParentMaritalStatus']).size().unstack()

pivot_df.plot(kind='bar', figsize=(12, 7), width=0.8)

plt.title("Ota-onalar ma'lumoti va oilaviy holati bo'yicha taqsimot", fontsize=14)
plt.xlabel("Ota-onaning ma'lumoti", fontsize=12)
plt.ylabel("Oila soni", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="Oilaviy holati")
plt.grid(axis='y', linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 7))

transport_counts = df['TransportMeans'].value_counts()
ax[0].pie(transport_counts, labels=transport_counts.index, autopct='%1.1f%%',
          colors=['#ffcc99', '#66b3ff'], startangle=90, explode=(0.05, 0))
ax[0].set_title("Talabalarning maktabga kelish transporti")

marital_counts = df['ParentMaritalStatus'].value_counts()
ax[1].bar(marital_counts.index, marital_counts.values, color='blue', edgecolor='black')
ax[1].set_title("Ota-onalarning oilaviy holati (Soni)")
ax[1].set_ylabel("Talabalar soni")

plt.tight_layout()
plt.show()
```

```python
# NaN qiymatlarni hisobga olmaslik uchun unique() dan keyin dropna() ishlatamiz
ordered_groups = sorted(df['EthnicGroup'].dropna().unique())

fig, ax = plt.subplots(1, 2, figsize=(15, 6))

# 1. Count Plot
sns.countplot(data=df, x='EthnicGroup', ax=ax[0], palette='viridis',
              order=ordered_groups)
ax[0].set_title('Ethnic Group bo\'yicha o\'quvchilar soni')

# 2. Pie Chart
group_counts = df['EthnicGroup'].value_counts()
ax[1].pie(group_counts, labels=group_counts.index, autopct='%1.1f%%',
          startangle=140, colors=sns.color_palette('viridis'))
ax[1].set_title('Ethnic Group ulushi (%)')

plt.tight_layout()
plt.show()
```
